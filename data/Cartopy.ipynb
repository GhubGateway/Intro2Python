Since we are using geospatial data, we not only **```import cartopy```** but also **```import cartopy.crs```**, which allows us to use the coordinate reference systems (CRS) class within Cartopy. 

````python
# import using abbreviations
import cartopy
import cartopy.crs as ccrs
````

## Making basic maps

We can use [**matplotlib**](Figures.ipynb) to make basic maps. We can follow the steps below:

1. Import necessary libraries: **matplotlib.pyplot, cartopy, cartopy.crs**

````python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
````

2. Set up the axes using the  **```axes()```** function from matplotlib. The argument will declare the projection for the map, which will be taken from the [Cartopy projection list](https://scitools.org.uk/cartopy/docs/latest/reference/projections.html#cartopy-projections). This will store an instance of the subclass **GeoAxes** from Cartopy, which allows us to <ins>add coastlines to our map</ins>.

````python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()

plt.show
````

This is just a general world map, but we can make a more complicated map (e.g. contour plots, add images to a map, vector plots).

3. Set up the axes using **```axes()```**.

````python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import numpy as np
import netCDF4 as nc

# Use Dataset() function from netCDF4 to open
dsst = nc.Dataset('sst_monthly.nc')

# Create arrays
sst = np.array(dsst['sst'][0,:,:])
sstlat = np.array(dsst['lat'])
sstlon = np.array(dsst['lon'])

# Set up access
ax = plt.axes(projection=ccrs.PlateCarree())
````

4. Use the **```contourf()```** function from **```matplotlib```** and input our arrays as our **x, y, and z**. We will also <ins>*transform*</ins> the data to our current map projection. Lastly, we'll **```show()```** our map.

````python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import numpy as np
import netCDF4 as nc

# Use Dataset() function from netCDF4 to open
dsst = nc.Dataset('sst_monthly.nc')

# Create arrays
sst = np.array(dsst['sst'][0,:,:]) # we put 0 since x is day which is not necessary here
sstlat = np.array(dsst['lat'])
sstlon = np.array(dsst['lon'])

# Set up access
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()

# Add data to map
plt.contourf(sstlon,sstlat,sst, transform = ccrs.PlateCarree())
plt.show
````

## Adding gridlines and labels

To add gridlines to a map, we can use **```gridlines()```**:



```python

gridlines(crs=None, draw_labels=False, xlocs=None, ylocs=None, dms=False, x_inline=None, y_inline=None, auto_inline=True, xformatter=None, yformatter=None, xlim=None, ylim=None, rotate_labels=None, xlabel_style=None, ylabel_style=None, labels_bbox_style=None, xpadding=5, ypadding=5, offset_angle=25, auto_update=False, formatter_kwargs=None, **kwargs)
```


The above are default values for the argument, but you can view each argument description [here](https://scitools.org.uk/cartopy/docs/latest/reference/generated/cartopy.mpl.geoaxes.GeoAxes.html#cartopy.mpl.geoaxes.GeoAxes).

````python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import numpy as np
import netCDF4 as nc

# Use Dataset() function from netCDF4 to open
dsst = nc.Dataset('sst_monthly.nc')

# Create arrays
sst = np.array(dsst['sst'][0,:,:])
sstlat = np.array(dsst['lat'])
sstlon = np.array(dsst['lon'])

# Set up access
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()

# Add gridlines
# draw_labels adds lat-lon labels to our gridlines
ax.gridlines(draw_labels=True)

# Add data to map
plt.contourf(sstlon,sstlat,sst, transform = ccrs.PlateCarree())
plt.show
````
